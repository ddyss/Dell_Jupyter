```python
import scipy.io
import torch
import torch.nn as nn
```

```python
dd=scipy.io.loadmat('F:/train/train_dataset/new_machine/subset_30_360_rand6/x1x1.mat')['x1x1']
```

```python
dd=torch.from_numpy(dd)
dd=dd.reshape(1,1,128,512)
# dd=dd.astype(np.float32)
dd=dd.type(torch.FloatTensor)
dd=dd.cuda()
```

```python
class DjgNet(nn.Module):
    def __init__(self):
        super(DjgNet,self).__init__()
        self.conv1 = nn.ConvTranspose2d(1,32,(3,4),(1,2),1) #
        self.conv2 = nn.ConvTranspose2d(32,64,(3,4),(1,2),1)
#         self.conv3 = nn.ConvTranspose2d(64, 1, 2, 2)
#         self.conv4 = nn.ConvTranspose2d(64, 32, 2, 2)
#         self.conv5 = nn.ConvTranspose2d(32, 1, 2, 1)
        self.relu=nn.ReLU(inplace=True)
    def forward(self,x):
        x=self.conv1(x)
        x=self.relu(x)
        x=self.conv2(x)
        x=self.relu(x)
#         x=self.conv3(x)
#         x=self.relu(x)
#         x=self.conv4(x)
#         x=self.relu(x)
#         x=self.conv5(x)
        return x
```

```python
djgnet=DjgNet()
djgnet=djgnet.to('cuda')
yy=djgnet(dd)
print(yy.shape)
```

```python
data=yy[0,30,:,:]
data.shape
```
