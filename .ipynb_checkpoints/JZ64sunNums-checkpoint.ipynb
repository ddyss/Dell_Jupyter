```python
import scipy.io
import h5py
import numpy as np
a=h5py.File('F:/colorbar/CNN2/train/real.60.25_70.mat','r')   #scipy.io.loadmat
b=h5py.File('F:/colorbar/CNN2/train/real.60.25_100.mat','r')   #h5py.File
da=[]
db=[]
data=[]  #列表   元组用（）
for i in range(60):
    da=[a[element[i]][:] for element in a['djg']]
    db.append(da)
for i in range(25):
    for j in range(60):
        data.append(db[j][i]) 
ta=[]
tb=[]
target=[]
for i in range(60):
    ta=[b[element[i]][:] for element in b['djg']]
    tb.append(ta)
for i in range(25):
    for j in range(60):
        target.append(tb[j][i]) 
        
```

```python
import torch.nn as nn
from torch.utils.data import Dataset
class MyDataset(Dataset):
    def __init__(self,data,target,transform):
        self.data=data
        self.target=target
        self.transform=transform
    def __getitem__(self,index):
        data=self.data[index]
        data=data.reshape(1,2030,128)
        data=data.astype(np.float32)
        data=data*1000
        target=self.target[index]
        target=target.astype(np.float32)
        target=target.reshape(1,2030,128)
        target=target*1000
        return data,target
    def __len__(self):
        return len(self.data)
from torchvision import transforms
transform=transforms.Compose([
    transforms.ToTensor()
])

dataset=MyDataset(data=data,target=target,transform=transform)
class DjgNet(nn.Module):
    def __init__(self):
        super(DjgNet,self).__init__()
        self.conv1=nn.Conv2d(1,32,5,1,2) #输入通道，输出通道，卷积核大小，步长，填充
        self.conv2=nn.Conv2d(32,64,5,1,2)
        self.conv3=nn.Conv2d(64,64,5,1,2)
        self.conv4=nn.Conv2d(64,32,5,1,2)
        self.conv5=nn.Conv2d(32,1,5,1,2)
        self.relu=nn.ReLU(inplace=True)
    def forward(self,x):
        x=self.conv1(x)
        x=self.relu(x)
        x=self.conv2(x)
        x=self.relu(x)
        x=self.conv3(x)
        x=self.relu(x)
        x=self.conv4(x)
        x=self.relu(x)
        x=self.conv5(x)
        return x
djgnet=DjgNet()
djgnet=djgnet.to('cuda')
from torch.utils.data import DataLoader
dataloader=DataLoader(dataset=dataset,batch_size=4,shuffle=True)
loss_func=nn.L1Loss()
from torch import optim
optimiazer=optim.Adam(djgnet.parameters())
for epoch in range(3):
    for i,(data,target)in enumerate(dataloader):
        data=data.cuda()
        target=target.cuda()
        out=djgnet(data)
        loss=loss_func(out,target)
        optimiazer.zero_grad()
        loss.backward()
        optimiazer.step()
        print(epoch,' ',i,':',loss.item())
import torch
torch.save(djgnet,'djgnet.60.4.pkl')
```
