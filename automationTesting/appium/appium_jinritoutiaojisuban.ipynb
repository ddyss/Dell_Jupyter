```python
from appium import webdriver
from appium.webdriver.extensions.android.nativekey import AndroidKey
from appium.webdriver.common.touch_action import TouchAction
from time import sleep

desired_caps = {
  'platformName': 'Android', # 被测手机是安卓
  'platformVersion': '10', # 手机安卓版本
  'deviceName': 'xiaomi 10', # 设备名，安卓手机可以随意填写
  'appPackage': 'com.ss.android.article.lite', # 启动APP Package名称
  'appActivity': '.activity.SplashActivity', # 启动Activity名称
#   'unicodeKeyboard': True, # 使用自带输入法，输入中文时填True
  'resetKeyboard': True, # 执行完程序恢复原来输入法
  'noReset': True,       # 不要重置App
  'newCommandTimeout': 6000,  #设置命令超时时间, 两条 appium 命令间的最长时间间隔，若超过这个时间，appium 会自动结束并退出app
#   'automationName' : 'UiAutomator2' #不设置会默认为UiAutomator1；这是安卓框架    iOS对应框架：XCUITest 框架
  # 'app': r'd:\apk\bili.apk',
}

# 启动APP；连接Appium Server，初始化自动化环境；默认的4723
driver = webdriver.Remote('http://localhost:4723/wd/hub', desired_caps)

# 设置隐式等待时间，开屏广告？
# driver.implicitly_wait(5)

sleep(5)

# 根据id定位‘三’标签处，点击
driver.find_element_by_id("com.ss.android.article.lite:id/agt").click()

# 选中体育，点击   ；但是好像还是不行啊     注意不要加elements，否则就变成列表了，没有click功能
# driver.find_element_by_xpath('/hierarchy/android.widget.FrameLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/android.widget.FrameLayout/android.widget.LinearLayout/com.ss.android.account.customview.slidingdrawer.SuperSlidingDrawer/android.widget.RelativeLayout[2]/android.widget.FrameLayout/android.widget.GridView/android.widget.RelativeLayout[15]/android.widget.TextView').click()

# 记得加 click()点击操作   https://my.oschina.net/u/4394125/blog/3310778
driver.find_element_by_android_uiautomator('new UiSelector().text("体育")').click() #通过text文本来定位
# driver.find_element_by_xpath("//*[contains(text(), '体育')]").click() #这个和下面这个还是不行，只有上面的能成功。
# driver.find_element_by_xpath("//android.widget.TextView[@text='体育']").click()

# action = TouchAction(driver)
# action.press(x=600, y=800).release().perform() # [554,736][789,846] 体育

window_width = driver.get_window_size()['width']
window_height = driver.get_window_size()['height']
        
def swipeDown(driver, t=500):
    '''向上滑动屏幕'''
    x1 = window_width * 0.5  # x坐标
    y1 = window_height * 0.5  # 起始y坐标
    y2 = window_height * 0.3  # 终点y坐标
    driver.swipe(x1, y1, x1, y2, t)
    
def swipeContentDown(driver, t=500):
    '''向上滑动屏幕'''
    x1 = window_width * 0.5  # x坐标
    y1 = window_height * 0.25  # 起始y坐标
    y2 = window_height * 0.2  # 终点y坐标
    driver.swipe(x1, y1, x1, y2, t)

for i in range(300):

    # 开始看新闻
    action = TouchAction(driver)
    action.press(x=window_width*0.5, y=window_height*0.5).release().perform()
    for i in range(10):
        sleep(2)
        swipeContentDown(driver) #Original error: Could not proxy command to the remote server. Original error: socket hang up


    # 返回
    driver.keyevent(4)
    swipeDown(driver)

# 这一句好像影响执行了，明明没有在for循环了啊，却只执行一次就退出来了，关闭后居然可以了。。。
# input('**** Press to quit..')#只有执行了这一句，才会执行下一句，关闭driver
# driver.quit()
```
