```python
'''
给一串数字，输出这些数字能够组成的合法的最大时间，数字不够则用0补全  LeetCode 949
输入：“274536”
输出：“23:57:46”
'''
```

```python
A = "245"
```

```python
# 时间复杂度： O(1)，实际上为n**3，这里n为6，一个常数216，还是可以接受的。 空间复杂度： O(1)。
import itertools
A = "245"
class Solution():
    def largestTimeFromDigits(self, A, base = 6):
        
        # 长度大于6，报错
        if len(A) > base:
            return False
        # 长度小于等于6，补0
        if len(A) <= base:
            A = "0" * (base-len(A)) + A

        # 字符串存入列表
        listStr = list(A)#['0', '0', '0', '2', '4', '5']
        res = []
        for lS in listStr:
            # 字符变整型
            res.append(int(lS))#[0, 0, 0, 2, 4, 5]
        
        # 从高到低排序，按要求填入，第一个出来的即为最大时间
        res.sort(reverse=True) #[5, 4, 2, 0, 0, 0]
        
        # permutations 返回可迭代对象的所有数学全排列方式
        for a,  b, c, d, e, f in itertools.permutations(res):
            # 判断时间合法与否
            if a*10+b <24 and c*10+d<60 and e*10+f<60:
                return f"{a}{b}:{c}{d}:{e}{f}"     #'20:54:00'
        return False

Solution().largestTimeFromDigits(A)
```
