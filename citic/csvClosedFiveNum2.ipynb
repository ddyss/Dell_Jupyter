```python
# 暴力for循环
# 删除异常数据，排序后先排除大于100的，减小复杂度
# 不用找到最好的最优解，考虑复杂度，考虑银行实际应用，比如找出5个最接近20的

res = []
base = 100
begin = float('inf')
for i in range(len(values)):
    for j in range(i+1,len(values)):
        for k in range(j+1,len(values)):
            for m in range(k+1,len(values)):
                for n in range(m+1,len(values)):
                    tmp = values[i]+values[j]+values[k]+values[m]+values[n]
                    res.append(abs(tmp-base))
                    
                    # 减少无畏的打印、计算；最后一次打印出来是就是最小的一组
                    if abs(tmp-base) < begin:
                        begin = min(abs(tmp - base),begin)
                        print(f'这五位客户索引位置：{i},{j},{k},{m},{n}')
                        print(f'这五位客户账户余额：{values[i]},{values[j]},{values[k]},{values[m]},{values[n]}')
                                          
```

```python
# https://leetcode-cn.com/problems/3sum-closest/        #LeetCode16  这是3个数求和
# solution/lei-si-shang-yi-ti-de-shuang-zhi-zhen-dai-ma-jian-/
nums = values
target = 100
class Solution:      
    def fiveSumClosest(self, nums, target):
        n=len(nums)
        if n==3: #只有三个数直接相加
            return nums[0]+nums[1]+nums[2]
        d=float('inf')
        nums=sorted(nums)
        ans=0
        for i in range(n):
            l,r=i+1,n-1
            while l<r:
                x=(nums[i]+nums[l]+nums[r])-target
                if abs(x)<d:#更新最短距离
                    d=abs(x)
                    ans=nums[i]+nums[l]+nums[r]
                if x>0: #大于target，所以需要更小的数，右指针左移
                    r-=1
                elif x<0:#小于target，所以需要更大的数，左指针右移
                    l+=1
                else:#与target相等，直接返回就行
                    return target
        return ans
```
