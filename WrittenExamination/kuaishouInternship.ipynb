```python
# 寻找两个有序数组的中位数。给定两个大小为m和n的有序数组nums1和nums2，合并有序数组并排序（不能用sort方法），
# 请你找出新数组的中位数（注：需要考虑两个数组为空的情况）
# LeetCode 4 寻找两个正序数组的中位数

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        newnum = self.merge_sort(nums1,nums2)
        n = len(newnum)
        if n % 2 == 1:
            return newnum[n//2]
        else:
            return (newnum[n//2] + newnum[n//2 - 1])/2
    

    
    def merge_sort(self,nums1,nums2):
        if not nums1:
            return nums2
        if not nums2:
            return nums1
        res = []
        i = 0
        j = 0
        while i < len(nums1) and j < len(nums2):
            if nums1[i] < nums2[j]:
                res.append(nums1[i])
                i += 1
            else:
                res.append(nums2[j])
                j += 1

        res.extend(nums1[i:])
        res.extend(nums2[j:])
        return res
        
    
```

```python
# vivo 笔试最后一个编程题，掷骰子，骰子不限次数，拆分数字
# 输入：3   输出：4
# 如总点数为3，则有4中情况：111,12,21,3

# 输入：4   输出：8
# 如总点数为4，则有6中情况：1111,112,121,211,13,31,22,4

# 最后罗列找规律，发现是前n项和，加上一个斐波那契数列
# 0 1 2 4 8 13 20 29

x = int(input())

class Solution:
    def feibo(self,x):
        # 先搞一个斐波那契数列
        if x < 2:
            return 0
        elif x == 2:
            return 1
        elif x == 3:
            return 1
        else:
            return self.feibo(x-1) + self.feibo(x-2)

    def countWay(self,x):        
        res = 0
        for i in range(x):
            res += i
        
        tmp = self.feibo(x)
        return res + tmp
    
print(Solution().countWay(x))
```
